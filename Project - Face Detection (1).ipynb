# FACE DETECTION

AIM: The aim is to create a model to detect the faces in a picture.

Here, we selected a picture of 14 people. We will train the model using this picture and then check the accuracy it is giving us.

```python
#Create and activate a new environment to work on this project and then later install all the necessary libraries.
```

```python
#Import the libraries
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the Image.
image = cv2.imread('C:\\Users\\Chandu\\Desktop\\projects\\people.jpg')
```

```python
#View the image in graph format.
plt.imshow(image)
plt.show()
```

```python
#Display the shape of the image.
display(image.shape)
```

```python
#print the image pixel details.
print(image)
```

```python
#Load the face detector xml file.
face_detector = cv2.CascadeClassifier('C:\\Users\\Chandu\\Desktop\\projects\\haarcascade_frontalface_default.xml')
face_detector
```

```python
#Check the number of faces detected using this algorithm.
detections = face_detector.detectMultiScale(image, scaleFactor = 1.03, minSize=(45,45), minNeighbors=4)
display (len(detections))
```

```python
#View the dimensions of the faces detected.
display (detections)
```

```python
#For the faces detected using the algorithm draw rectangle faces for each of them.
for (x, y, w, h) in detections:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0,0,0), 2)
cv2.imshow('',image)
cv2.waitKey(0)
```

# Observation:
This model was able to detect 13 faces out of the 14 faces in the image.
